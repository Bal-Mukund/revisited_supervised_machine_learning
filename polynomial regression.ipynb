### without feature engineering

```python
# creating data
x = np.arange(0,20,1)
y = x**2 + 1

# converting this 1-D array into an 2-d array
X = x.reshape(-1,1)
```

```python
w_val, b_val, cost_val = run_gradient_descent(X,y, iterations=1000, alphas= 1e-2)
```

```python
X.shape
```

```python
value = np.dot(w_val,x)
```

```python
# model
m = X.shape[0]
values = np.zeros(m)

for i in range(m):
    model = np.dot(w_val, X[i]) + b_val
    values[i] += model
```

```python
plt.scatter(x, y, marker='x', c='r', label="Actual Value")
plt.plot(x, values, label="Predicted Value")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Without Feature Engineering")
plt.legend()
plt.show()
```

### Feature Engineering

```python
# creating data
x = np.arange(0,20,1)
y = x**2 + 1

X = x**2 
# converting this 1-D array into an 2-d array
X = X.reshape((20,1))
```

```python
w_val, b_val, cost_val = run_gradient_descent(X,y,iterations=10000, alphas=1e-5)
```

```python
y_hat = (w_val* X) + b_val
y_hat
```

```python
# plotting  
plt.scatter(x, y, marker='x',c='r', label= 'Actual value')
plt.plot(x, y_hat, label='predicted value')
plt.title("feature engineering")
plt.xlabel("x")
plt.ylabel('y')
plt.legend()
plt.show()
```

### experimenting more feature engineering

```python
x = np.arange(0,20,1)
y_val = x**2 + 1

# adding more feature engineering 
# numpy .c_[] adds an array along second axis (column)
X = np.c_[x, x**2, x**3]
```

```python
w_val, b_val, cost_val = run_gradient_descent(X, y_val, iterations=10000, alphas=1e-7)
```

```python
m,n = X.shape
y_hat = np.zeros((m,n))
for i in range(m):
    y = np.dot(w_val, X[i, :]) + b_val
    y_hat[i] += y 
```

```python
x.shape
y_hat
```

```python
plt.scatter(x,y_val, marker='x', c='r', label= "Actual Values")
plt.plot(x, y_hat, label="Predicted Values")
plt.xlabel('x')
plt.ylabel('y')
plt.title("experimenting more feature engineering")
plt.legend()
plt.show()
```
