### normalizing data using in built sklearn z-score normalization

```python
scaler = StandardScaler()
z_norm = scaler.fit_transform(X_train)
```

### creating model using sklearn

```python
sgdr = SGDRegressor(max_iter=1000)
sgdr.fit(z_norm, y_train)
```

### viewing parameters

```python
w = sgdr.coef_
b = sgdr.intercept_
```

```python
w, b
```

### making predictions

```python
# using sklearn 
y_hat = sgdr.predict(z_norm)
y_hat.shape
```

```python
# using numpy
m = z_norm.shape[0]
predicted_y = np.zeros(m)
for i in range(m):
    predicted_y[i] = np.dot(w, z_norm[i]) + b
    
```

### plotting 

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(12,3), sharey=True)

n = z_norm.shape[1]
for i in range(n):
    ax[i].scatter(X_train[:, i], y_train, label="Actual")
    ax[i].scatter(X_train[:, i], y_hat, label="Predicted")
    ax[i].set_xlabel(X_features[i])

ax[0].set_ylabel("Price")
ax[0].legend()

fig.suptitle("Original price vs sklearn Predicted Price")
plt.show()
```
